```python
Y1 = data.to_numpy()
```

```python
import nltk
nltk.download('wordnet')
documents = []

from nltk.stem import WordNetLemmatizer

stemmer = WordNetLemmatizer()

for sen in range(0, len(Y1)):
    
    # Remove all the special characters
    document = re.sub(r'\W', ' ', str(Y1[sen]))
    # remove numbers
    document = re.sub(r'\d+', '', document)
    
    # remove all single characters
    document = re.sub(r'\s+[a-zA-Z]\s+', ' ', document)
    
    # Remove single characters from the start
    document = re.sub(r'\^[a-zA-Z]\s+', ' ', document) 
    
    # Substituting multiple spaces with single space
    document = re.sub(r'\s+', ' ', document, flags=re.I)
    
    # Removing prefixed 'b'
    document = re.sub(r'^b\s+', '', document)
    
    # Converting to Lowercase
    document = document.lower()
    
    # Lemmatization
    document = document.split()

    document = [stemmer.lemmatize(word) for word in document]
    document = ' '.join(document)
    
    documents.append(document)
```

```python
def listToString(documents): 
    
    # initialize an empty string
    str1 = "" 
    
    # traverse in the string  
    for ele in documents: 
        str1 += ele  
    
    # return string  
    return str1 
```

```python
text = listToString(documents)
```

```python
from nltk.tokenize import sent_tokenize, word_tokenize
nltk.download('punkt')
token_text = word_tokenize(text)
```

```python

import nltk
from nltk.corpus import webtext
from nltk.probability import FreqDist
 
wt_words = token_text
data_analysis = nltk.FreqDist(wt_words)
 
# Let's take the specific words only if their frequency is greater than 50.
filter_words = dict([(m, n) for m, n in data_analysis.items() if len(m) > 10])
 
for key in sorted(filter_words):
    print("%s: %s" % (key, filter_words[key]))
 
data_analysis = nltk.FreqDist(filter_words)
```

```python
data_analysis.plot(30, cumulative=False, title = "Most Frequent words in Strivers", label= True, color="black")
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
## Creating FreqDist for whole BoW, keeping the 20 most common tokens
all_fdist = FreqDist(data_analysis).most_common(30)

## Conversion to Pandas series via Python Dictionary for easier plotting
all_fdist = pd.Series(dict(all_fdist))

## Setting figure, ax into variables
fig, ax = plt.subplots(figsize=(17,6))

## Seaborn plotting using Pandas attributes + xtick rotation for ease of viewing
all_plot = sns.barplot(x=all_fdist.index, y=all_fdist.values, ax=ax, color="grey")
plt.title("Strivers Content key features")
plt.xticks(rotation=30);
```
